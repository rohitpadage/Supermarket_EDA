```python
df1['hr']=df1.Time.str.split(':').str[0]
#hr=df1.Time.str.split(':').str[0]
#minute=df1.Time.str.split(':').str[1]

```

```python
df1.head()
```

```python
count_mor=count_after=count_eve=0

for i in hr:
    if int(i)<12:
        count_mor+=1
    elif int(i)>=12 and int(i)<18:
        count_after+=1
    else:
        count_eve+=1
print(count_mor)
print(count_after)
print(count_eve)
```

```python
df1.head()
```

```python
df1['Time_Zone'] = df1.apply(lambda _: '', axis=1)
df1.head()
```

```python
for x in df1.index:
    h=df1.loc[x, "hr"]
    if int(h)<12:
        #count_mor+=1
        df1.loc[x, "Time_Zone"]="Morning"
    elif int(h)>=12 and int(h)<18:
        #count_after+=1
        df1.loc[x, "Time_Zone"]="Afternoon"
    else:
       # count_eve+=1
        df1.loc[x, "Time_Zone"]="Evening"
```
